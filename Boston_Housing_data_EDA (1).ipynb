# Data Manipulation

*   Add a new column to the DataFrame that categorizes the ‘medv’ (median value of owner-occupied homes) column into ‘Low’, ‘Medium’, and ‘High’.



```python
data['MEDV'].max() # maximum value of MEDV
```

```python
data['MEDV'].min() # minimum value of MEDV
```

```python
data["MEDV STATUS"] = pd.cut(data['MEDV'],bins = [0,20,35,50],labels=["Low","Medium","High"]) # new columns MEDV STATUS contain Low, Medium, High.
```

```python
data
```

Replace the ‘chas’ column (Charles River dummy variable) with ‘Yes’ if chas = 1 and ‘No’ if chas = 0.

```python
data["CHAS"]=data['CHAS'].map({1:"Yes",0:"No"}) # Mapping 1 to "Yes" and 0 to "No."
data
```

Rename the column ‘rm’ to ‘rooms’.

```python
data.rename(columns={"RM":"ROOMS"},inplace= True) # rename rm to rooms.
```

```python
data
```

# NumPy Operations:
Convert the ‘age’ column (proportion of owner-occupied units built prior to 1940) from the DataFrame to a NumPy array and perform the following operations:


*   Compute the mean and standard deviation.
*   Normalize the array.





```python
age = data["AGE"] # use age columns
np.array(age)
```

```python
np.mean(age) # mean of age column
```

```python
np.std(age) # standard deviation of age column
```

```python
# normalized age column using max min normalization
normalized_age = (age - np.min(age)) / (np.max(age) - np.min(age))
normalized_age
```
